7- Un sistema de transiciones es una generalización de los autómatas no determinísticos con $\varepsilon$-transiciones adicionales con palabras de longitud o igual a 1. Dé una definición formal de los sistemas de transiciones, y pruebe que los lenguajes aceptados por un sistema de transiciones son aceptados por un DFA.<br><br>
El autómata que representa al sistema de transiciones puede hacer una transición a uno o varios estados a partir de un símbolo $a$ o una secuencia (palabra) $w$, donde $\vert w \vert \geq 1$. Un sistema de transiciones se representa igual que un $\varepsilon$-NFA con una excepción: la función de transición tiene que incluir información sobre transiciones para una palabra w. El sistema de transiciones $S_t$ se representa por: <br><br>

$S_t = (Q_{St}, \sum, \hat{\delta}, q_0, F)$<br><br>

en el cual todos los componentes tienen la misma interpretación que en un $\varepsilon$-NFA, excepto que la función de transición $\delta$ pasa a ser ahora $\hat{\delta}$, una función de transición con los argumentos:<br>

<ol type="1">
  <li>un estado q y, </li>
  <li>un elemento de $\sum\ast \cup \lbrace \varepsilon \rbrace$ es decir, cualquier palabra $w\in \sum\ast$ tal que $\vert w \vert \geq 1$  o el símbolo $\varepsilon$ (palabra vacía), $\varepsilon$ no puede ser elemento de $\sum$ .</li>
</ol> <br>

Esta función de transición $\hat{\delta}$ puede procesar tanto símbolos como palabras, pues puede ser de la forma:
<ol type="a">
  <li> $\hat{\delta}(q,w)$, donde $\vert w \vert = 1$, lo que sería equivalente a procesar un único símbolo ó,  </li> 
  <li> $\hat{\delta}(q,w)$ = $\hat{\delta}(q,xa)$ donde $x$ es una palabra y $a$ es el último símbolo de ésta. Esta función es equivalente a la función de transición extendida definida para un NFA donde: $\hat{\delta}(q,w)=\delta(\hat{\delta}(q,x),a)$</li>
</ol> 

En el sistema de transiciones que estamos definiendo, esto se representa de la siguiente forma: estando en un estado $q$, al observar una transición saliente de $q$ etiquetada con $w$ se pasa a uno o varios estados a los que llega dicha transición.

9 - Respecto al uso de autómatas (máquinas de estado finito) para el diseño de personajes de videojuegos:<br>
- ¿Cuáles son las ventajas del uso de la abstracción de máquinas de estado frente a una implementación utilizando condicionales?<br><br>
Cuando se decide modelar un personaje para un videojuego, la implementación de su clase en código puede incluir una gran cantidad de métodos y funciones; cuando se decide modelar el comportamiento del personaje a partir de simples sentencias condicionales, se puede generar una gran complejidad en la integridad y eficacia del código, ya que no se está guardando memoria de diversas acciones que ha desarrollado el personaje para llegar a su estado actual, ni se establece una estrategia productiva para que estando en dicho estado el personaje pueda pasar a uno o varios estados después, a partir de acciones en el juego. La implementación de máquinas de estado permite dar solución a problemas de robustez y complejidad en el modelamiento de un personaje, pues permite establecer estados que son capaces de guardar memoria e incluir transiciones entre los mismos a partir de acciones llevadas a cabo por el personaje en el videojuego.
<br><br>
- ¿Cómo se utilizan los autómatas en el diseño de personajes para videojuegos?<br><br>
Los autómatas son empleados para definir los estados y el paso entre estados de un determinado personaje de un videojuego. Los componentes de dichos autómatas son dos básicamente: los estados y las características de funcionamiento de dichos estados. Los estados consisten de tres cosas: una prioridad, un enunciado booleano para determinar si el estado debe "ejecutarse" y la lógica que contiene el estado. Las características de funcionamiento del autómata son básicamente las siguientes: 1. Determinar los estados que puede alcanzar el jugador en un momento del juego, 2. Determinar la prioridad que tiene cada estado dentro del comportamiento del personaje, 3. Determinar cuando un estado necesita ser alcanzado, 4. Si el estado debe alcanzarse por el personaje, ejecutar la lógica definida para éste, y 5. Si el personaje no debe llegar a un estado determinado, buscar el estado correcto a partir de una acción. <br><br>
De dicha forma se pueden modelar los autómatas que controlen el comportamiento del personaje y su paso entre distintos estados a partir de las acciones que se requieran para alcanzar los mismos.<br><br>
