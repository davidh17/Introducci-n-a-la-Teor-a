 
Un ISBN es un código normalizado internacional para libros que actualmente consta de 13 dígitos. Los ISBN se calculan utilizando una fórmula matemática específica e incluyen un dígito de control que valida el código. (Tomado de https://www.isbn-international.org)

Cada ISBN se compone de 5 elementos separados entre sí por un espacio o un guion. Tres de los cinco elementos pueden variar en longitud:
<ul style="list-style-type:disc">
  <li>Elemento prefijo: actualmente sólo pueden ser 978 o 979. Siempre tiene 3 dígitos de longitud.</li>
  <li>Elemento de grupo de registro: identifica a un determinado país, una región geográfica o un área lingüística que participan en el sistema ISBN. Este elemento puede tener entre 1 y 5 dígitos de longitud.</li>
  <li>Elemento del titular: identifica a un determinado editor o a un sello editorial. Puede tener hasta 7 dígitos de longitud.</li>
  <li>Elemento de publicación: identifica una determinada edición y formato de un determinado título. Puede ser de hasta 6 dígitos de longitud.</li>
  <li>Dígito de control: es siempre el último y único dígito que valida matemáticamente al resto del número. Se calcula utilizando el sistema de Módulo 10 con pesos alternativos de 1 y 3
</ul>
